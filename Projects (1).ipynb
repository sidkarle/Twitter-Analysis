```python
text= '''The book is structured around the metaphor of a staircase, where each step
represents a critical element of reaching personal and professional
success. The first step, according to Ziglar, is Self-Image. He emphasizes
the importance of having a positive self-image and how it is foundational
to achieving anything meaningful in life.
The next step is building good Character, followed by the significance of
having the right Attitude. Ziglar suggests that a positive attitude
enables you to face challenges head-on and is instrumental in achieving
success.
Work and effort are also major themes in the book. Ziglar emphasizes that
setting Goals is another critical step on the staircase. Without goals,
according to him, it's easy to drift aimlessly through life. He encourages
the reader to write down their short-term and long-term goals and devise
action plans to achieve them.
Interpersonal relationships and creating a healthy environment by being
around positive, uplifting people is another point that Ziglar covers. The
book delves into various aspects of human psychology and social
interactions that can either elevate you or bring you down.
Lastly, Ziglar discusses Desire, the final step on the staircase. He
argues that without a burning desire to achieve your goals, all the other
steps are inconsequential. Ziglar doesn't just present abstract ideas; he
provides practical advice and actionable steps one can take to start
climbing their personal staircase to the top.'''
```

```python
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
print(nltk.download('stopwords'))
```

```python
word_count= {}
for word in text.split():
  word = word.lower()
  if word in word_count:
    word_count[word]+=1
  else:
    word_count[word] = 1
print(word_count)
```

```python
count=[]
for i in text.split():
  count.append(i)
print(len(count))
```

```python
count= sum(word_count.values())
count
```

```python
stop_words = set(stopwords.words('english'))
print(f"The number of stopwords: {len(stop_words)}")
print(stop_words)
stop_words.update(['book','effort'])
print(f"The number of stopwords: {len(stop_words)}")
print(stop_words)
```

```python
a=[]
for i in text.split():
  if i not in stop_words:
    a.append(i)
print(a)
b=" "
b.join(a)
print(b)
print(len(b))


```

```python
f= " ".join([word for word in text.split() if word.lower() not in stop_words])
f
```

```python
import string
st= ""
for i in f:
  if i not in string.punctuation:
    st+=i

st
```

```python
word_count= {}
for word in st.split():
  word = word.lower()
  if word in word_count:
    word_count[word]+=1
  else:
    word_count[word] = 1
print(word_count)
```

```python
count= sum(word_count.values())
count
```

```python
plt.figure(figsize=(10,8))
plt.imshow(WordCloud(background_color="orange").generate(st))
plt.axis("off")
plt.show()
```

```python
tweet_data= dataframe.drop(['user_name', 'location'],axis=1)
```

```python
tweet_data.head()
```

```python
import re
```

```python
def collect_hashtag(tweet):
  hastag = re.findall(r"#[a-zA-Z0-9]+",tweet)
  return ", ".join(hastag)
```

```python
tweet_data['hashtag']= tweet_data['tweet'].apply(lambda x : collect_hashtag(x))
```

```python
tweet_data.sample(5)
```

```python
def collect_mention(tweet):
  mention = re.findall(r"@[a-zA-Z0-9]+",tweet)
  return ", ".join(mention)
```

```python
tweet_data['mention'] = tweet_data['tweet'].apply(lambda x : collect_mention(x))
```

```python
tweet_data.sample(5)
```

```python
def count_hashtag(tweet):
  hashtag = re.findall(r"#[a-zA-Z0-9]+", tweet)
  return len(hashtag)
```

```python
tweet_data['count_hashtag']= tweet_data['tweet'].apply(lambda x : count_hashtag(x))
```

```python
tweet_data.sample(5)
```

```python
count= tweet_data['count_hashtag'].sum()
print(f"number of hashtag {count}")
```

```python
def count_mention(tweet):
  mention = re.findall(r"@[a-zA-Z0-9]+",tweet)
  return len(mention)
```

```python
all_hashtag = tweet_data['hashtag'].to_string()
hashtags= re.findall(r"#\w+",all_hashtag)
hashtags
```

```python
hashtag_dict = {}
for word in hashtags:
  if word in hashtag_dict:
    hashtag_dict[word] +=1
  else:
    hashtag_dict[word] =1
hashtag_dict
```

```python
sorted_hashtags= sorted(hashtag_dict.items(), key=lambda x: x[1], reverse= True)
sorted_hashtags
```

```python
top_six = sorted_hashtags[:5]
top_six
```

```python
hashtag, count = zip(*top_six)
print(hashtag)
print(count)
```

```python
import matplotlib.pyplot as plt
plt.bar(hashtag,count)
plt.show
```

sentiment analysis

```python
from  textblob import TextBlob
```

```python
def measure_sentiment(tweet):
  blob = TextBlob(tweet)
  return blob.sentiment.polarity
```

```python
tweet_data['sentiment_score'] = tweet_data['tweet'].apply(lambda x: measure_sentiment(x))
```

```python
tweet_data.head()
```

```python
def sentiment_label(score):
  if score> 0:
    return "positive"
  elif score<0:
    return "negative"
  else:
    return "neutral"

```

```python
tweet_data['sentiment_label']= tweet_data['sentiment_score'].apply(lambda x: sentiment_label(x))
```

```python
tweet_data.head()
```

```python
count = tweet_data['sentiment_label'].value_counts()
print(count)
```

```python
count.plot(kind= 'pie',autopct='%1.1f%%')
plt.title("product evaluation")
plt.show()
```
